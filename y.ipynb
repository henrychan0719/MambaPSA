```python
import torch

print("=" * 60)
print("PyTorch / CUDA")
print("=" * 60)

print("torch:", torch.__version__)
print("cuda available:", torch.cuda.is_available())

if torch.cuda.is_available():
    print("gpu:", torch.cuda.get_device_name(0))
    print("cuda version:", torch.version.cuda)

print("\n" + "=" * 60)
print("Mamba packages")
print("=" * 60)

try:
    import mamba_ssm
    print("mamba_ssm OK")
    print("version:", getattr(mamba_ssm, "__version__", "unknown"))
except Exception as e:
    print("mamba_ssm FAIL")
    print(e)

try:
    import causal_conv1d
    print("causal_conv1d OK")
except Exception as e:
    print("causal_conv1d FAIL")
    print(e)

try:
    from mamba_ssm.ops.selective_scan_interface import selective_scan_fn
    print("selective_scan_fn OK")
except Exception as e:
    print("selective_scan_fn FAIL")
    print(e)

print("\n" + "=" * 60)
print("Mamba forward test")
print("=" * 60)

try:
    from mamba_ssm import Mamba

    device = "cuda" if torch.cuda.is_available() else "cpu"

    model = Mamba(
        d_model=64,
        d_state=16,
        d_conv=4,
        expand=2,
    ).to(device)

    x = torch.randn(2, 128, 64).to(device)
    y = model(x)

    print("Mamba forward OK")
    print("input shape :", x.shape)
    print("output shape:", y.shape)
    print("device:", y.device)

except Exception as e:
    print("Mamba forward FAIL")
    print(e)
```
